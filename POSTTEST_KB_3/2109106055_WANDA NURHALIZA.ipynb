### MENGIMPORT LIBRARY DAN MENAMPILKAN DATASET

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tv_shows.csv")
df.info()
```

### GRAFIK PIE

##### Dapat kita lihat disini bahwa yang menonton film dengan kategori usia 7+ ada 25.6%, 13+ ada 0.3%, 16+ ada 30.7%, 18+ ada 26.3% dan semua usia ada 17%.

```python
df = df.dropna(subset=['Age'])
```

```python
data_pie = df['Age'].value_counts().rename_axis('Age').reset_index(name='Id_count')
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    data_pie.Id_count, 
    labels     = data_pie.Age, 
    startangle = 90,
    autopct    = '%.1f%%'
)
plt.title('Data yang Menonton Film berdasarkan Kategori Usia')
plt.show()
```

### GRAFIK BATANG

##### Semua usia menonton dari semua tahun tayang.

```python
x = list(df.iloc[:,1])
y = list(df.iloc[:,1])
```

```python
sns.barplot(
    data = df,
    x    = 'Age', 
    y    = 'Year',
)
plt.title('Perbandingan yang Menonton Film berdasarkan Tahun Tayang dan Kategori Usia')
plt.show()
```

### SCATTER PLOT

##### Orang yang menonton film di Disney+ pada tahun 1980an - 2022 lebih banyak daripada tahun 1980 ke bawah.

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10,10))
sns.scatterplot(
    data  = df,
    x     = 'Disney+',  
    y     = 'Year'
)
plt.title('Orang yang Menonton Film di Disney+ berdasarkan Tahun Tayang')
plt.show()
```

### GRAFIK BOXPLOT

##### Keterangannya disini adalah 0  berarti tidak ada yang menonton, sedangkan 1 berarti ada yang menonton. Disini lebih banyak yang tidak ada yang menonton karena ini adalah data dari 1900an. Sedangkan, netflix belum ada di tahun tersebut.

```python
plt.figure(figsize=(10,20))
sns.boxplot(
    data   = df,
    x      = 'Netflix',
    y      = 'Year',
)

plt.xlabel('Netflix')
plt.ylabel('Year')
plt.title('Orang yang Menonton Film di Netflix berdasarkan Tahun Tayang')
plt.show()
```
